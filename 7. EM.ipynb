```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the Iris dataset
data = pd.read_csv('iris.csv')  # Replace 'iris_dataset.csv' with the path to your Iris dataset

# Extract the features
X = data[['sepal_length', 'sepal_width', 'petal_length','petal_width']].values

# Number of data points
N = X.shape[0]

# Number of clusters
K = 3

# Randomly initialize the cluster means and covariances
np.random.seed(0)
cluster_means = np.random.randn(K, X.shape[1])
cluster_covariances = np.tile(np.eye(X.shape[1]), (K, 1, 1))

# Initialize the cluster priors
cluster_priors = np.ones(K) / K

# E-step: Compute the responsibilities
def compute_responsibilities(X, cluster_means, cluster_covariances, cluster_priors):
    responsibilities = np.zeros((X.shape[0], K))
    for n in range(N):
        for k in range(K):
            diff = X[n] - cluster_means[k]
            exponent = -0.5 * np.dot(np.dot(diff.T, np.linalg.inv(cluster_covariances[k])), diff)
            responsibilities[n, k] = cluster_priors[k] * (1 / (np.sqrt(np.linalg.det(cluster_covariances[k])))) * np.exp(exponent)
    responsibilities /= np.sum(responsibilities, axis=1, keepdims=True)
    return responsibilities

# M-step: Update the cluster means, covariances, and priors
def update_parameters(X, responsibilities):
    Nk = np.sum(responsibilities, axis=0)
    cluster_means = np.dot(responsibilities.T, X) / Nk[:, np.newaxis]
    for k in range(K):
        diff = X - cluster_means[k]
        cluster_covariances[k] = np.dot((diff * responsibilities[:, k][:, np.newaxis]).T, diff) / Nk[k]
    cluster_priors = Nk / N
    return cluster_means, cluster_covariances, cluster_priors

# Run the EM algorithm
num_iterations = 100

for i in range(num_iterations):
    # E-step
    responsibilities = compute_responsibilities(X, cluster_means, cluster_covariances, cluster_priors)
    
    # M-step
    cluster_means, cluster_covariances, cluster_priors = update_parameters(X, responsibilities)

# Assign each data point to the cluster with the highest responsibility
labels = np.argmax(responsibilities, axis=1)

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.xlabel('Sepal and Petals')
plt.ylabel('Flower Type')
plt.title('GMM Clustering of Iris Dataset')
plt.show()
```
